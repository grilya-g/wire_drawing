```python
import numpy as np
import pandas as pd

# Получение датафреймов с компонентами тензора напряжений
df1 = df_dict[0]  # σ1
df2 = df_dict[1]  # σ2
df3 = df_dict[2]  # σ3

# Определяем имена столбцов с компонентами напряжений (например, r_1, r_2, ...)
stress_cols = [col for col in df1.columns if col.startswith('r_')]
param_cols = [col for col in df1.columns if not col.startswith('r_')]

# Извлекаем значения компонент напряжений
s1 = df1[stress_cols]
s2 = df2[stress_cols]
s3 = df3[stress_cols]

# Рассчитываем интенсивность напряжений по формуле Мизеса
von_mises = np.sqrt(0.5 * ((s1 - s2) ** 2 + (s2 - s3) ** 2 + (s3 - s1) ** 2))

# Собираем датафрейм с параметрами и интенсивностями напряжений
df_von_mises = df1[param_cols].copy()
df_von_mises[stress_cols] = von_mises

# Просмотр результата
df_von_mises.head()
```
